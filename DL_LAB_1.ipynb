Import tensorflow library

```python
import tensorflow as tf
```

create tensors

```python
# Create a scalar (0-D tensor)#
a=tf.constant(10)
print(a)
# Create a vector (1-D tensor)
v=tf.constant([1,2,34,3])
print(v)
# Create a matrix (2-D tensor)
m=tf.constant([[1,2],[3,4]])
print(m)
# Create a 3-D tensor
t3=tf.constant([[[1,2],[3,4]],[[5,6],[7,8]]])
print(t3)
```

```python
# Addition of tensors
a=tf.constant([10,20,30])
b=tf.constant([1,23,34])
c=tf.add(a,b)
print(c)
# Subtraction of tensors
c=tf.subtract(a,b)
print(c)
# Element-wise multiplication
c=tf.multiply(a,b)
print(c)
# Division of tensors
c=tf.divide(a,b)
print(c)
```
